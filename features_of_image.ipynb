```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/content/Hrithik_Roshan_4.webp')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_float = np.float32(gray_image)
corners = cv2.cornerHarris(gray_float, 2, 3, 0.04)
image[corners > 0.01 * corners.max()] = [0, 0, 255]

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Harris Corner Detection')
plt.show()
```

```python
import cv2
from skimage.feature import graycomatrix, graycoprops # Changed 'greycomatrix' to 'graycomatrix' and 'greycoprops' to 'graycoprops'
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('/content/Hrithik_Roshan_4.webp', cv2.IMREAD_GRAYSCALE)

# Compute the GLCM
glcm = graycomatrix(image, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)

# Compute GLCM properties
contrast = graycoprops(glcm, 'contrast')[0, 0]
dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]
homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
energy = graycoprops(glcm, 'energy')[0, 0]

# Print GLCM properties
print(f'Contrast: {contrast}')
print(f'Dissimilarity: {dissimilarity}')
print(f'Homogeneity: {homogeneity}')
print(f'Energy: {energy}')

# Visualize the GLCM
plt.imshow(glcm[:, :, 0, 0], cmap='gray')
plt.title('GLCM (0 degrees)')
plt.show()
```

```python
import cv2
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/Hrithik_Roshan_4.webp')
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Extract HOG features and HOG image
hog_features, hog_image = hog(
    gray_image,
    orientations=9,
    pixels_per_cell=(8, 8),
    cells_per_block=(2, 2),
    block_norm='L2-Hys',
    visualize=True,
    multichannel=False
)

# Print the shape of the HOG features
print('HOG feature vector shape:', hog_features.shape)

# Rescale the HOG image for better display
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

# Display the original and HOG images
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(gray_image, cmap=plt.cm.gray)
ax1.set_title('Input Image')

ax2.axis('off')
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)
ax2.set_title('HOG Image')

plt.show()
```

```python
import cv2
import numpy as np
from skimage.feature import local_binary_pattern
import matplotlib.pyplot as plt

def convert_to_lbp(image_path, radius=1, n_points=8):
    # Read the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if image is None:
        print("Error: Image not found or unable to read.")
        return None

    # Compute LBP
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')

    return lbp

def plot_lbp(original_image, lbp_image):
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    ax[0].imshow(original_image, cmap='gray')
    ax[0].set_title('Original Image')
    ax[0].axis('off')

    ax[1].imshow(lbp_image, cmap='gray')
    ax[1].set_title('LBP Image')
    ax[1].axis('off')

    plt.show()

# Example usage
# Use the correct image path that exists and is accessible.
image_path = '/content/Hrithik_Roshan_4.webp' # Changed to the path from the Global Variables
lbp_image = convert_to_lbp(image_path)

if lbp_image is not None:
    original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    plot_lbp(original_image, lbp_image)
```
